## SVR




### Importing Libraries 

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
```

### Importing Dataset

```python
dataset = pd.read_csv('Position_Salaries.csv')

X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

```python
print(X)
```

```python
print(y)
```

```python
y = y.reshape(len(y), 1)
```

```python
print(y)
```

### Feature Scaling

SVR model will not work at all without feature scaling

```python
from sklearn.preprocessing import StandardScaler

sc_x = StandardScaler()
sc_y = StandardScaler()
X = sc_x.fit_transform(X)
y = sc_y.fit_transform(y)
```

```python
print(X)
```

```python
print(y)
```

### Training the SVR Model on the whole dataset

```python
from sklearn.svm import SVR

regressor = SVR(kernel='rbf')

regressor.fit(X, y)
```
