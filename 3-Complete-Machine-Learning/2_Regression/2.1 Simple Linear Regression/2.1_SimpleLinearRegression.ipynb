### Simple Linear Regression

#### Importing the Libraries

```python
# Importing Libraries 
import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd
```

#### Importing the dataset

```python
dataset = pd.read_csv('Salary_Data.csv')

X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
X
```

#### Splitting the dataset into training data and test data

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

#### Training the Simple Linear Regression Model on the training data

```python
from sklearn.linear_model import LinearRegression

regressor = LinearRegression()

# The fit method will train the model on the training dataset
regressor.fit(X_train, y_train)
```

#### Predict the Test Set Results

```python
y_pred = regressor.predict(X_test)
```

#### Visualizing the training set results

```python
plt.scatter(X_train, y_train, color='r')
plt.plot(X_train, regressor.predict(X_train))
plt.title('Salaray vs Experience plot (Training Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

#### Visualizing the Test Set Resuls

```python
plt.scatter(X_test, y_test, color='r')
plt.plot(X_test, regressor.predict(X_test))
plt.title('Salaray vs Experience plot (Test Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```
